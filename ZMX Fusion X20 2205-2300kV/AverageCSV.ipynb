```python
import pandas as pd
import numpy as np

csv1 = "withLum5x4x3_2017-04-18_123723.csv"
csv2 = "withLum5x4x3_2017-04-18_144611.csv"
csv3 = "withLum5x4x3_2017-04-18_145028.csv"

csvOut = "withLum5x4x3_AVG.csv"

data1 = pd.read_csv(csv1)
data2 = pd.read_csv(csv2)
data3 = pd.read_csv(csv3)

l1, l2, l3 = len(data1.index), len(data2.index), len(data3.index)

maxIndex = l1
if(l1 > l2):
    maxIndex = l2
    if(l2 > l3):
        maxIndex = l3
        
data1 = data1[:maxIndex]
data2 = data2[:maxIndex]
data3 = data3[:maxIndex]
headers = list(data1)

seriesList = []
for i in range(0, maxIndex):
    vec1 = data1[i:i+1]
    vec2 = data2[i:i+1]
    vec3 = data3[i:i+1]
    vec_avg = []
    for col in headers:
        s = float(vec1[col]) + float(vec2[col]) + float(vec3[col])
        avg = s/3
        vec_avg.append(avg)
    seriesList.append(pd.Series(vec_avg, index=headers))
data = pd.DataFrame(seriesList, columns=headers)

data.to_csv(csvOut)
```
